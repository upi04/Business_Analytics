```python
import pandas as pd
import numpy as np

# Mengimpor data
data = pd.read_csv('online_shoppers_intention.csv')

# Mengisi nilai null di kolom 'ProductRelated_Duration' dengan rata-rata
data['ProductRelated_Duration'] = data['ProductRelated_Duration'].fillna(data['ProductRelated_Duration'].mean())

# Menghapus baris dengan nilai null di kolom 'BounceRates'
data = data.dropna(subset=['BounceRates'])
data.head()
```

```python
# Mengonversi kolom 'Month' menjadi format yang konsisten (misalnya, huruf kecil)
data['Month'] = data['Month'].str.lower()

# Mengonversi kolom 'Revenue' menjadi boolean (True/False)
data['Revenue'] = data['Revenue'].astype(bool)

data.head()
```

```python
# Mengganti nilai negatif di kolom 'ProductRelated_Duration' dengan 0
data['ProductRelated_Duration'] = data['ProductRelated_Duration'].apply(lambda x: max(x, 0))

# Memeriksa apakah ada nilai negatif di kolom 'Administrative_Duration'
data['Administrative_Duration'] = data['Administrative_Duration'].apply(lambda x: max(x, 0))

# Menampilkan beberapa baris data setelah penanganan nilai negatif
data.head()
```

```python
# Menghapus baris duplikat
data = data.drop_duplicates()

# Menampilkan jumlah baris sebelum dan sesudah penghapusan duplikasi
print(f"Jumlah baris sebelum duplikasi: {len(data)}")
data = data.drop_duplicates()
print(f"Jumlah baris setelah duplikasi dihapus: {len(data)}")
```

```python
# Mengonversi kolom 'Month' menjadi nama bulan yang konsisten
data['Month'] = data['Month'].str.capitalize()

# Menampilkan hasil konversi kolom 'Month'
data.head()
```

```python
# Membuat kolom baru untuk mengelompokkan durasi sesi
data['SessionDurationGroup'] = pd.cut(data['ProductRelated_Duration'], bins=[0, 60, 120, 180, 300, float('inf')], labels=['Short', 'Medium', 'Long', 'Very Long', 'Extremely Long'])

# Menampilkan beberapa baris data dengan fitur baru
data.head()
```

```python
# Menyimpan dataset yang telah dibersihkan ke file CSV
output_file_path = r'D:\projetc_sql\cleaned_online_shoppers_intention.csv'
data.to_csv(output_file_path, index=False)

output_file_path
```
